```python
from docx import Document
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

doc_path = 'dbscan.docx'
doc = Document(doc_path)

text = [p.text for p in doc.paragraphs if p.text]

data = [line.split(',') for line in text]
df = pd.DataFrame(data, columns=['X', 'Y', 'Z', 'Cluster'], dtype=np.float64)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

clusters = sorted(df['Cluster'].unique())
colors = plt.cm.rainbow(np.linspace(0, 1, len(clusters)))

for color, cluster in zip(colors, clusters):
    cluster_data = df[df['Cluster'] == cluster]
    ax.scatter(cluster_data['X'], cluster_data['Y'], cluster_data['Z'], color=color, label=f'Cluster {int(cluster)}')

ax.set_xlabel('Xing')
ax.set_ylabel('Wei')
ax.set_zlabel('Guijing')
ax.set_title('Clustering')
ax.legend()

plt.show()
```
